UK Energy Transition: Analysing the Shift from Fossil Fuels to Renewables (1990–2020).

This project analyses UK energy consumption data from 1990 to 2020, examining the shift from fossil fuels to renewable sources across 30 years of national energy records.

The following questions are to be answered using the dataset:

•At what point did renewable consumption exceed a meaningful threshold relative to fossil fuels, and how long did it take to get there from near-zero?

•Which renewable sources are driving the growth; is it one dominant type pulling the total up, or is growth spread across multiple sources?

•Are there years where fossil fuel consumption dropped sharply, and do those years correspond to renewable growth or something else entirely (recession, policy change, price spike)?

Calculate renewable share as a percentage of total energy consumption for each year to answer question 1.

```python
df["renewable_share"] = (df["Energy from renewable & waste sources"] / df["Total energy consumption of primary fuels and equivalents"]) * 100
```

```python
df[['Year', 'renewable_share']].head(10)
```

```python
df[df['renewable_share'] >= 10]
```

Filter to find the first year renewables exceeded 10% of total energy consumption, a commonly cited policy benchmark.

```python
first_year = df[df['renewable_share'] >= 10]['Year'].iloc[0]
start_year = df['Year'].iloc[0]
years_taken = first_year - start_year

print(f"Renewables first reached 10% of total energy consumption in {first_year}")
print(f"This took {years_taken} years from {start_year}")
```

It took the UK 28 years (1990–2018) to reach the often-cited milestone of 10% renewable energy, and this level was only sustained in the final three years of the dataset (2018–2020). Given the UK’s target of near-total decarbonisation by 2050, this trajectory indicates that the pace of transition must accelerate significantly.

While this analysis doesn’t establish causation, several plausible factors help explain the slow progress. Technological limitations played a role, key sources like solar photovoltaics and liquid biofuels were not widely operational until the mid-2000s. In addition, political prioritisation likely influenced progress, with long-term energy investments often deprioritised during economic downturns in favour of short-term economic support, sometimes increasing reliance on fossil fuels and reducing funding for renewables.

Identify the renewable source columns only, excluding year, total renewable and fossil fuel columns, to analyse individual source contributions.

Sum each renewable source across all years to compare cumulative contributions and identify dominant sources.

```python
renewable_cols = df.columns[3:]
df[renewable_cols].sum().sort_values(ascending=False)
```

This is the total energy created by diffrerent forms of renewables. Most people associate the renewable transition with wind and solar, but the data says biomass burning (wood) has been the largest single contributor since 1990. However a source could rank highly just because it's been around longest, not because it's driving recent growth. Wood has been used since the start of this dataset. Solar barely existed until recently but may be growing fastest. So the real question is: which sources are accelerating? see below.

Compare first and last year values for each renewable source to identify which sources have grown most over the period.

```python
first_row = df[renewable_cols].iloc[0]
last_row = df[renewable_cols].iloc[-1]
growth = last_row - first_row
growth.sort_values(ascending=False)
```

Visualise cumulative growth by source as a horizontal bar chart to make dominance immediately visible.

```python
import matplotlib.pyplot as plt

growth.sort_values().plot(kind='barh', figsize=(10,8))
plt.title('Growth in renewable energy by source (first vs last year)')
plt.xlabel('Growth in PJ')
plt.tight_layout()
plt.show()
```

Wind,wave,tidal jumps to the top when you look at growth rather than cumulative contribution. Wood drops to second and solar, which barely registered in the cumulative ranking, is now sixth above landfill gas which had been ahead of it cumulatively. That demonstrates that solar is a recent accelerator, not a long-term contributor. Growth isn't spread evenly wind,wave,tidal energy are clearly the dominant factor behind recent expansion while Bio energy is a significant and steady resource whether in a cumulative or growth setting. The UK is an island nation with access to the north sea which happens to also be one of the windiest places on the planet. The Uk is leaning into this strength which is demonstarted in the wind,tidal,wave ranking above. However this growth calculation compares endpoint values only. It does not capture intermediate volatility or the shape of growth over time. A source that grew sharply then declined would not be distinguished from one that grew steadily. The line chart below captures intermediate trajectory.

Plot all renewable sources as lines over time to capture intermediate trajectory and identify sources that grew steadily vs those that peaked and plateaued.

```python
df.plot(x='Year', y=list(renewable_cols), figsize=(12,8))
plt.title('Renewable energy sources over time')
plt.ylabel('Energy (PJ)')
plt.tight_layout()
plt.show()
```

This graph captures intermediate volatility missed by the endpoint calculation. Unlike the previous chart which only compared start and end values, this reveals the shape of growth over time.

Wind, wave and tidal only recently overtook wood in terms of energy production after wood plateaued around 2015. Wood remains the second highest contributor in absolute terms but its stalled growth means the renewable story is increasingly dependent on wind, wave and tidal as it represents close to 50% of total renewable energy production. This could point to a concentration risk if deployment slows due to planning, grid capacity or policy constraints. Biodiesel is actively declining, unusual for a renewable source, implying policy changes, cost pressures or substitution. Solar continues its ascent but appears to be plateauing in the final five years of the data, unsurprising given the UK's limited sunlight compared to more solar-optimal geographies.

Renewables have been growing strongly across the period but the pace needs to accelerate substantially to meet 2050 net zero targets. The current trajectory is encouraging but wind dependency and the wood plateau represent structural questions about whether that acceleration is achievable.

Calculate year on year change in fossil fuel to identify years of sharp decline

```python
df["fossil_fuel_change"] = df["Total energy consumption of primary fuels and equivalents"].diff()
df[["Year", "fossil_fuel_change"]]
```

Calculate year on year change in fossil fuel and renewable consumption to identify years of sharp decline and whether they correspond to renewable growth or demand suppression.

```python
df["renewable_change"] = df["Energy from renewable & waste sources"].diff()
df[["Year", "fossil_fuel_change", "renewable_change"]]
```

Visualise year on year changes for both fossil fuels and renewables on the same chart to compare volatility and trajectory across the full period.

```python
df.plot(x='Year', y=['fossil_fuel_change', 'renewable_change'], figsize=(12,6))
plt.title('Year on year change: fossil fuels vs renewables')
plt.ylabel('Change in PJ')
plt.axhline(y=0, color='black', linestyle='--', linewidth=0.8)
plt.tight_layout()
plt.show()
```

The orange renewable line is remarkably stable and consistently above zero; renewables grew almost every single year regardless of what fossil fuels were doing. The blue fossil fuel line is volatile, swinging dramatically in both directions, and has a clear downward drift from around 2005 onwards. 2020 is visually unmistakable. Notice how the fossil fuel line behaves differently before and after roughly 2005. Before 2005 it oscillates around zero with no clear direction. Post 2005 it spends most of its time below zero with occasional recoveries. That's not just economic shocks, that's a structural change in energy consumption. That's a really clear picture. Look at the four key years:

2009: fossil fuels -14.412, renewables +0.795

2011: fossil fuels -13.249, renewables +0.906

2014: fossil fuels -11.282, renewables +2.094

2020: fossil fuels -23.061, renewables +1.601

The pattern is consistent across all four; fossil fuels drop sharply while renewables continue growing. That tells you the fossil fuel drops are not being driven by renewable substitution, because renewable growth is too small to account for the scale of the decline. Something else is suppressing total energy demand.

2020 is by far the largest single-year drop and this is almost certainly caused by the covid pandemic. 2009 is the aftermath of the 2008 financial crisis. 2011 was characterised by a summer stock market crash, high volatility, and fears of a "double-dip" recession and was primarily caused by a combination of the U.S. debt ceiling crisis and the worsening European sovereign debt crisis. 2014 was primarily caused by a dramatic, 40% to 50% collapse in global oil prices between June and December, driven by a surge in US shale oil production and weakening demand from China and the Eurozone. The data shows a correlation between economic shocks and fossil fuel drops but not definitive causation.

Secondly look at the renewable growth in the drop years compared to the surrounding years. 2014 shows +2.094, and 2013 shows +1.715, which are among the strongest renewable growth years in the dataset. Renewables were accelerating regardless of what fossil fuels were doing. This suggests that renewables aren't as impacted by external factors like economic downturns or global contexts like pandemics and wars. This is another argument for further investment in renewables. Fuel costs have been a major point of conversation in recent times, with prices increasing excessively due to circumstances like the Ukraine and iran wars. Developing renewables as a main source of energy could bring stability to the energy sector and the cost of living as a whole.

Summary of key findings:

1)Slow early progress toward renewables

It took the UK 28 years (1990–2018) to reach just 10% renewable energy, and only maintained that level in the final three years of the dataset. This highlights how slow the transition has been and suggests major acceleration is needed to hit 2050 net zero targets, especially given past limits like immature technology and shifting political priorities.

2)What’s really driving renewable growth

While biomass (wood) has been the biggest overall contributor historically, it’s now plateaued, meaning it’s no longer driving growth. The real expansion is coming from wind, wave, and tidal, now responsible for about half of renewable energy, with solar emerging quickly but starting to level off. This creates a growing reliance on wind, raising potential risks if that sector slows.

3)Renewables vs fossil fuels

Renewables show steady, consistent growth every year, while fossil fuels are volatile and declining, especially after 2005. During major economic shocks (e.g. 2009, 2014, 2020), fossil fuel use drops sharply but not because renewables replace them, rather due to reduced overall demand. This suggests renewables are more stable and resilient, strengthening the case for expanding them to improve energy and price stability.
